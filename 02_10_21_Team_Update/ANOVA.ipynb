```python
import pandas as pd
import statsmodels.api as sa
import statsmodels.formula.api as sfa
from statsmodels.stats.multicomp import pairwise_tukeyhsd

df = pd.read_csv (r'C:\Users\leey2\Downloads\combined_dysphagaid.csv')
#change path upon download

lm = sfa.ols('StarValue ~ C(Product)', data=df).fit()
anovastar = sa.stats.anova_lm(lm)
print(anovastar)
#significant difference between product star means

tukey = pairwise_tukeyhsd(endog=df['StarValue'], groups=df['Product'], alpha=0.05)
print(tukey)
#post-hoc analysis

lm = sfa.ols('polarity ~ C(Product)', data=df).fit()
anova_polarity = sa.stats.anova_lm(lm)
print(anova_polarity)
#significant difference between polarity

lm = sfa.ols('subjectivity ~ C(Product)', data=df).fit()
anova_subjectivity = sa.stats.anova_lm(lm)
print(anova_subjectivity)
#significant difference between subjectivity
```
